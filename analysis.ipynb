```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
ml_jobs = pd.read_csv('data\\mlengineer_cleaned.csv')
```

```python
languages = ['python', 'sql', 'c\+\+', 'java', 'scala']
values = []
total = len(ml_jobs)
for language in languages:
    popularity = ml_jobs['Job Description'].str.contains(language, case=False)
    popularity = popularity.sum() / total * 100
    values.append(popularity)
    
ax = sns.barplot(x=values, y=languages, palette="Blues_d", 
                order=[language for _,language in sorted(zip(values,languages), reverse=True)])
ax.bar_label(ax.containers[0],fmt="%.0f%%")
plt.show()
```

```python
qualifications = ['Bachelor','master','PhD']
values = []
total = len(ml_jobs)
for qualification in qualifications:
    popularity = ml_jobs['Job Description'].str.contains(qualification, case=False)
    popularity = popularity.sum() / total * 100
    values.append(popularity)
    
ax = sns.barplot(x=values, y=qualifications, palette="Blues_d", 
                order=[qualification for _,qualification in sorted(zip(values,qualifications), reverse=True)])
ax.bar_label(ax.containers[0],fmt="%.0f%%")
plt.show()
```
